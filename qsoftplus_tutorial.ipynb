```python
import torch
import torch.nn as nn
from q_softplus import qSoftplus, ShiftedqSoftplus

import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-4, 4, steps=1000)
```

```python
q = 0.5
qs = qSoftplus(q=q)
y = qs(x)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x.numpy(), y.detach().numpy(), label=f"qSoftplus q={q}")

plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, ncol=1)
plt.show()
```

```python
q = 0.5
sqs = ShiftedqSoftplus(q=q)
y = sqs(x)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x.numpy(), y.detach().numpy(), label=f"qSoftplus q={q}")

plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, ncol=1)
plt.show()
```
